```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
df=pd.read_csv("D:\\Saurabh\\Colab Files\\Artificial_Neural_Networks Colab\\Churn_Modelling.csv")
```

```python
df.head()
```

```python
#independent and dependent 
X=df.iloc[:,3:-1]
y=df.iloc[:,-1]
print("X: Independent_Varibale")
print("y: Dependent_Varibale")
```

```python
#shape
print('X_shape:',X.shape)
print('y_shape:',y.shape)
```

```python
#get dummies
geo =pd.get_dummies(X['Geography'],drop_first=True)
gen =pd.get_dummies(X['Gender'],drop_first=True)
```

```python
print(geo)
print(gen)
```

```python
#concat the geo and gen
X =pd.concat([X,geo,gen],axis=1)
```

```python
X.head()
```

```python
#drop Geography and Gender
X.drop(labels=['Geography','Gender'],axis=1,inplace=True)
```

```python
X.head(3)
```

```python
#scaling 
from sklearn.preprocessing import StandardScaler
Scaler =StandardScaler()
X_scale =Scaler.fit_transform(X)
```

```python
#train_test_splits
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X_scale,y,test_size=0.3,random_state=11)
print('X_train_shape:',X_train.shape)
print('y_train_shape:',y_train.shape)
print('X_test_shape:',X_test.shape)
print('y_test_shape:',y_test.shape)
```

```python
#creating ANN
classifer = Sequential()
```

```python
#adding input layers to ANN that is first layers
classifer.add(Dense(units=15,kernel_initializer='he_uniform',activation='relu',input_dim=11))
#2nd_layers
classifer.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
#3rd_layer
classifer.add(Dense(units=25,kernel_initializer='he_uniform',activation='relu'))
#output_layers 
classifer.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
#summary_classifer
classifer.summary()
```

```python
#compile the classifier
classifer.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_his=classifer.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
#predicts for test_data
y_pred =classifer.predict(X_test)
y_pred =(y_pred>0.5)
```

```python
from sklearn import metrics
print("Accuracy_Score :",metrics.accuracy_score(y_test,y_pred))
print('Classification_report :\n',metrics.classification_report(y_test,y_pred))
print('Confusion_matrix :\n',metrics.confusion_matrix(y_test,y_pred))
```
